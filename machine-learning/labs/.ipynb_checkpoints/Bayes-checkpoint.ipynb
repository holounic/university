```python
import pandas as pd
import jax
import jax.numpy as jnp
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score as accuracy, roc_curve as roc, precision_score as precision
from sklearn.model_selection import train_test_split
import os
import re

DATA_PATH = './data/messages'
```

## Load data

```python
LEGIT_RE = re.compile(r'legit')
SPAM_RE = re.compile(r'spmsg')


def load_part(path):
    files = os.listdir(path)
    X, y = [], [] 
    for f in files:
        if LEGIT_RE.search(f):
            y.append(0)
        elif SPAM_RE.search(f):
            y.append(1)
        else:
            raise Exception()
        with open(os.path.join(path, f)) as file:
            lines = file.readlines()
        text_idx = 1 if len(lines) == 2 else 2
        text = [int(x) for x in lines[text_idx].split(' ')[:-1]]
        header = [int(x) for x in lines[0].split(' ')[1:-1]]
        X.append({'words': text + header, 'header': None})
    return X, y

def load_all():
    folds = os.listdir(DATA_PATH)
    return list(zip(*([load_part(os.path.join(DATA_PATH, fold)) for fold in folds])))
```

```python
DATA_X, DATA_Y = load_all()
DATA_Y = jnp.array(DATA_Y)
```

## N-gram transformation

```python
def to_n_gram(n):
    def split(text):
        return [tuple(text['words'][i : i + n]) for i in range(len(text['words']) - n)]
    return split

def texts_to_n_gram(texts, y, n):
    f = to_n_gram(n)
    result = []
    for (text, y_dot) in zip(texts, y):
        for x in f(text):
            result.append((x, y_dot))
    return result
```

## Naive Bayes

```python
def get_probs(n_grams):
    probs_pos, probs_neg = {}, {}
    for (features, y) in n_grams:
        cur_probs = probs_pos if y == 1 else probs_neg
        for i, feature in enumerate(features):
            prev_entries = cur_probs.setdefault(feature, [0 for _ in range(len(features))])
            prev_entries[i] += 1
    return probs_pos, probs_neg

def bayes(X, y, lam, n):
    aprior = [1 - sum(y) / len(y), sum(y) / len(y)] # 0 = not spam, 1 = spam
    n_grams = texts_to_n_gram(X, y, n)
    probs_positive, probs_negative = get_probs(n_grams)
    n_gram_eval = to_n_gram(n)
    
    def classifier_func(n_gram, y):
        cur_probs = probs_positive if y == 1 else probs_negative
        feature_probs = [cur_probs.get(word, [0 for _ in range(n)])[i] for (i, word) in enumerate(n_gram)]
        return jnp.log(lam[y] * aprior[y]) + sum([jnp.log(p) for p in feature_probs]) 
    
    def classify_n_gram(n_gram):
        prob_pos = classifier_func(n_gram, 1)
        prob_neg = classifier_func(n_gram, 0)
        return 1 if prob_pos > prob_neg else 0
                      
    def classify_one(text):
        n_grams = n_gram_eval(text)
        predictions = [classify_n_gram(x) for x in n_grams]
        positive_prob = sum(predictions) / len(predictions)
        return jnp.array((1 - positive_prob, positive_prob)) # first = not spam prob, second = spam prob  
    
    def classifier(texts, in_probs=True):
        probs = jnp.array([classify_one(text) for text in texts])
        return probs if in_probs else jnp.argmax(probs, axis=-1)
    
    return classifier
```

```python
model = bayes(DATA_X[0], DATA_Y[0], lam=(0.9, 0.9), n=2)
model(DATA_X[1][:4], in_probs=False) == DATA_Y[1][:4]
```

## Train & Fun

```python
def build_train_test(data_x, data_y, test_idx):
    train_x = []
    train_y = jnp.array([])
    for i in range(len(data_x)):
        if i == test_idx:
            continue
        train_x = train_x + data_x[i]
        train_y = jnp.concatenate([train_y, data_y[i]])
    return (train_x, train_y), (data_x[test_idx], data_y[test_idx]) # train_set, test_set
```

```python
def test_params(lams, ns):
    for lam in lams:
        print(f'Lambda: {lam}')
        for n in ns:
            print(f'::N: {n}')
            for test_idx in range(1):
                print(f'::::Fold with test part {test_idx}')
                
                (train_x, train_y), (test_x, test_y) = build_train_test(DATA_X, DATA_Y, test_idx)
                model = bayes(train_x, train_y, lam=lam, n=n)
                
                result = accuracy(model(test_x, in_probs=False), test_y)
                print(f'::::::Accuracy: {result}')
```

```python
test_params(((.01, .01), (.1, .1), (.5, .5)), (1, 3))
```

## Plot ROC curve

```python
(train_x, train_y), (test_x, test_y) = build_train_test(DATA_X, DATA_Y, 0)
model = bayes(train_x, train_y, lam=(0.01, 0.01), n=3)
```

```python
prediction = model(test_x, in_probs=False)
```

```python
fpr, tpr, _ = roc(test_y, prediction)
```

```python
plt.title('ROC Curve')
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.plot(fpr, tpr)
```

## Find lambda for no false positives

```python
# precision = tp / (tp + fp) => fp = 0 <=> precision = 1 

def test_legit(lam_spam, lams_legit, n):
    accuracies = []
    for lam in lams_legit:
        (train_x, train_y), (test_x, test_y) = build_train_test(DATA_X, DATA_Y, 0)
        model = bayes(train_x, train_y, lam=(lam, lam_spam), n=n)
        prediction = model(test_x, in_probs=False)
        acc = accuracy(prediction, test_y)
        print(f'Accuracy: {acc} Precision: {precision(test_y, prediction)} with lambda legit: {lam}')
        accuracies.append(acc)
    return accuracies
```

```python
lam_spam = 0.01
best_n = 3

lams_legit = jnp.linspace(0.01, .25, num=5)
```

```python
accuracies = test_legit(lam_spam, lams_legit, best_n)
accuracies
```

```python
plt.title('Lambda_legit and accuracy')
plt.xlabel('lambda_legit')
plt.ylabel('Accuracy')
plt.plot(lams_legit, accuracies)
```
